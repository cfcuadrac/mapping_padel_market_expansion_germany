```python
import pandas as pd

# Assuming your merged dataframe is already loaded
# This recreates the key columns you showed

# Step 1: Calculate POTENTIAL PADEL PLAYERS
# Based on conversion rates: 50% tennis, 25% gym, 15% football, 10% new

def calculate_opportunity_index(df):
    """
    Calculate opportunity index for padel expansion
    
    Inputs needed in df:
    - tennis_players
    - gym_members
    - football_players
    - padel_court_count
    - trend_score_five_years
    - population
    """
    
    # Calculate potential padel players (weighted by conversion likelihood)
    df['potential_padel_players'] = (
        df['tennis_players'] * 0.50 +      # 50% from tennis
        df['gym_members'] * 0.25 +         # 25% from gym/fitness
        df['football_players'] * 0.15 +    # 15% from football
        df['population'] * 0.10 * 0.01     # 10% newcomers (assume 1% of population)
    ).round().astype(int)
    
    # Step 2: Calculate DEMAND SCORE (0-100)
    # Normalize potential players to 0-100 scale
    max_potential = df['potential_padel_players'].max()
    df['demand_score'] = (
        (df['potential_padel_players'] / max_potential) * 100
    ).round(1)
    
    # Step 3: Calculate SUPPLY SCORE (0-100)
    # More courts = higher supply = LOWER opportunity
    # We invert this: fewer courts per capita = higher score
    
    # Courts per 100k population
    df['courts_per_100k'] = (
        (df['padel_court_count'] / df['population']) * 100000
    ).round(2)
    
    # Invert: cities with fewer courts per capita get higher scores
    max_density = df['courts_per_100k'].max()
    df['supply_gap_score'] = (
        100 - ((df['courts_per_100k'] / max_density) * 100)
    ).round(1)
    
    # Step 4: Calculate INTEREST SCORE (already 0-100 from Google Trends)
    df['interest_score'] = df['trend_score_five_years']
    
    # Step 5: Calculate FINAL OPPORTUNITY INDEX
    # Weighted average:
    # - 40% Demand (potential players)
    # - 40% Supply Gap (room for growth)
    # - 20% Interest (search trends)
    
    df['opportunity_index'] = (
        df['demand_score'] * 0.40 +
        df['supply_gap_score'] * 0.40 +
        df['interest_score'] * 0.20
    ).round(1)
    
    # Step 6: Rank cities
    df['rank'] = df['opportunity_index'].rank(ascending=False, method='min').astype(int)
    
    return df


# Example usage with your data:
# Recreate your merged dataframe structure
data = {
    'city': ['Hamburg', 'Berlin', 'München', 'Frankfurt am Main', 'Köln', 
             'Düsseldorf', 'Kiel', 'Stuttgart'],
    'state': ['Hamburg', 'Berlin', 'Bavaria', 'Hesse', 'North Rhine-Westphalia',
              'North Rhine-Westphalia', 'Schleswig-Holstein', 'Baden-Württemberg'],
    'population': [1851596, 3902645, 1604771, 775109, 1024621, 658245, 251379, 604804],
    'disposable_income': [29620, 26209, 31525, 28826, 27754, 27754, 29395, 30242],
    'trend_score_five_years': [100, 68, 47, 54, 71, 71, 48, 48],
    'padel_court_count': [79, 67, 52, 29, 25, 20, 7, 5],
    'tennis_court_count': [710, 909, 331, 224, 439, 350, 106, 280],
    'tennis_players': [33208, 161997, 27790, 13458, 19226, 11361, 4370, 11099],
    'gym_members': [340000, 523800, 430000, 100000, 150000, 80000, 17500, 90000],
    'football_players': [165794, 809436, 138743, 67190, 95986, 56719, 21815, 55410]
}

df = pd.DataFrame(data)

# Calculate opportunity index
df_scored = calculate_opportunity_index(df)

# Display results sorted by opportunity
result_columns = [
    'rank', 'city', 'opportunity_index', 'demand_score', 
    'supply_gap_score', 'interest_score', 'potential_padel_players',
    'padel_court_count', 'courts_per_100k'
]

print("="*80)
print("PADEL EXPANSION OPPORTUNITY INDEX")
print("="*80)
print(df_scored[result_columns].sort_values('opportunity_index', ascending=False).to_string(index=False))

print("\n" + "="*80)
print("TOP 3 CITIES FOR EXPANSION")
print("="*80)

top_3 = df_scored.nlargest(3, 'opportunity_index')
for idx, row in top_3.iterrows():
    print(f"\n#{int(row['rank'])} - {row['city'].upper()}")
    print(f"   Opportunity Index: {row['opportunity_index']}/100")
    print(f"   Potential Players: {row['potential_padel_players']:,}")
    print(f"   Current Courts: {row['padel_court_count']}")
    print(f"   Courts per 100k: {row['courts_per_100k']}")
    print(f"   Why: ", end="")
    
    reasons = []
    if row['demand_score'] > 70:
        reasons.append("high demand")
    if row['supply_gap_score'] > 70:
        reasons.append("low court density")
    if row['interest_score'] > 70:
        reasons.append("strong search interest")
    
    print(", ".join(reasons) if reasons else "balanced opportunity")

# Save results
df_scored.to_csv('opportunity_index_results.csv', index=False)
print("\n" + "="*80)
print("Results saved to 'opportunity_index_results.csv'")
print("="*80)
```

```python
"""
Calculate Padel Opportunity Index for German Cities
Uses actual data from your merged dataframe
Author: Claudio
"""

import pandas as pd

# Fix city name encoding issues
padel_tennis_counts["city"] = padel_tennis_counts["city"].replace({
    "Munich": "München",
    "Frankfurt": "Frankfurt am Main"
})

# Merge all data together
merged = (
    pop_income_state
    .merge(germany_gtrend, on="state", how="left")
    .merge(padel_tennis_counts, on="city", how="left")
    .merge(gym_members, on="city", how="left")
    .merge(city_football_tennis, on="city", how="left")
)

print("Data loaded successfully!")
print(f"Analyzing {len(merged)} German cities\n")

# ============================================================================
# STEP 2: Calculate Opportunity Index
# ============================================================================

def calculate_opportunity_index(df):
    """
    Calculate opportunity index based on:
    - Potential padel players (from tennis, gym, football, newcomers)
    - Supply gap (courts per capita)
    - Search interest (Google Trends)
    
    CORRECTED Conversion rates (% who will TRY padel):
    - 15% of tennis players will try padel
    - 12% of gym members will try padel
    - 7% of football players will try padel
    - 2% of general population (newcomers)
    
    Note: These rates reflect realistic adoption, not the composition
    of the padel player base.
    """
    
    # Make a copy to avoid modifying original
    df = df.copy()
    
    # Fill missing values with 0 to avoid errors
    df['tennis_players'] = df['tennis_players'].fillna(0)
    df['gym_members'] = df['gym_members'].fillna(0)
    df['football_players'] = df['football_players'].fillna(0)
    df['padel_court_count'] = df['padel_court_count'].fillna(0)
    df['trend_score_five_years'] = df['trend_score_five_years'].fillna(0)
    
    # -------------------------------------------------------------------------
    # A) DEMAND SCORE: Calculate potential padel players
    # -------------------------------------------------------------------------
    
    # More realistic conversion rates:
    # - 15% of tennis players likely to try padel (mid-range estimate)
    # - 12% of gym members likely to try padel (mid-range estimate)
    # - 7% of football players likely to try padel (mid-range estimate)
    # - 2% of general population as newcomers
    
    df['potential_padel_players'] = (
        df['tennis_players'] * 0.15 +      # 15% of tennis players
        df['gym_members'] * 0.12 +          # 12% of gym members
        df['football_players'] * 0.07 +     # 7% of football players
        df['population'] * 0.02              # 2% newcomers from general pop
    )
    
    # Round and convert to int (now safe because we filled NaN)
    df['potential_padel_players'] = df['potential_padel_players'].round(0).astype(int)
    
    # Normalize to 0-100 scale (handle case where all values are 0)
    max_potential = df['potential_padel_players'].max()
    if max_potential > 0:
        df['demand_score'] = (
            (df['potential_padel_players'] / max_potential) * 100
        ).round(1)
    else:
        df['demand_score'] = 0
    
    # -------------------------------------------------------------------------
    # B) SUPPLY GAP SCORE: Fewer courts per capita = higher opportunity
    # -------------------------------------------------------------------------
    
    # Calculate courts per 100k population (avoid division by zero)
    df['courts_per_100k'] = (
        (df['padel_court_count'] / df['population']) * 100000
    ).replace([float('inf'), float('-inf')], 0).round(2)
    
    # Invert the score: fewer courts = higher opportunity
    max_density = df['courts_per_100k'].max()
    
    if max_density > 0:
        df['supply_gap_score'] = (
            100 - ((df['courts_per_100k'] / max_density) * 100)
        ).round(1)
    else:
        df['supply_gap_score'] = 100  # If no courts exist, max opportunity
    
    # -------------------------------------------------------------------------
    # C) INTEREST SCORE: Google Trends (already 0-100)
    # -------------------------------------------------------------------------
    
    df['interest_score'] = df['trend_score_five_years']
    
    # -------------------------------------------------------------------------
    # D) FINAL OPPORTUNITY INDEX (weighted average)
    # -------------------------------------------------------------------------
    
    # Weights:
    # - 40% Demand (potential players)
    # - 40% Supply Gap (room for new courts)
    # - 20% Interest (search trends)
    
    df['opportunity_index'] = (
        df['demand_score'] * 0.40 +
        df['supply_gap_score'] * 0.40 +
        df['interest_score'] * 0.20
    ).round(1)
    
    # Rank cities by opportunity
    df['rank'] = df['opportunity_index'].rank(ascending=False, method='min').astype(int)
    
    return df


# ============================================================================
# STEP 3: Apply the calculation
# ============================================================================

df_final = calculate_opportunity_index(merged)

# ============================================================================
# STEP 4: Display results
# ============================================================================

# Select columns for display
result_columns = [
    'rank',
    'city',
    'opportunity_index',
    'demand_score',
    'supply_gap_score',
    'interest_score',
    'potential_padel_players',
    'padel_court_count',
    'courts_per_100k',
    'population',
    'disposable_income'
]

# Sort by opportunity index
results = df_final[result_columns].sort_values('opportunity_index', ascending=False)

print("="*90)
print("PADEL EXPANSION OPPORTUNITY INDEX - GERMAN CITIES")
print("="*90)
print(results.to_string(index=False))

# ============================================================================
# STEP 5: Show detailed insights for top 3 cities
# ============================================================================

print("\n" + "="*90)
print("TOP 3 EXPANSION OPPORTUNITIES - DETAILED ANALYSIS")
print("="*90)

top_3 = df_final.nlargest(3, 'opportunity_index')

for idx, row in top_3.iterrows():
    print(f"\n{'='*90}")
    print(f"RANK #{int(row['rank'])} - {row['city'].upper()}")
    print(f"{'='*90}")
    print(f"  Opportunity Index:        {row['opportunity_index']}/100")
    print(f"")
    print(f"  📊 SCORES:")
    print(f"     Demand Score:          {row['demand_score']}/100")
    print(f"     Supply Gap Score:      {row['supply_gap_score']}/100")
    print(f"     Interest Score:        {row['interest_score']}/100")
    print(f"")
    print(f"  👥 MARKET SIZE:")
    print(f"     Population:            {row['population']:,}")
    print(f"     Potential Players:     {row['potential_padel_players']:,}")
    print(f"     Disposable Income:     €{row['disposable_income']:,}")
    print(f"")
    print(f"  🏓 CURRENT SUPPLY:")
    print(f"     Existing Courts:       {row['padel_court_count']}")
    print(f"     Courts per 100k:       {row['courts_per_100k']}")
    print(f"")
    print(f"  💡 WHY THIS CITY:")
    
    # Generate insights
    reasons = []
    if row['demand_score'] >= 80:
        reasons.append(f"Very high demand ({int(row['potential_padel_players']):,} potential players)")
    elif row['demand_score'] >= 60:
        reasons.append(f"Strong demand ({int(row['potential_padel_players']):,} potential players)")
    
    if row['supply_gap_score'] >= 80:
        reasons.append("Severely undersupplied with courts")
    elif row['supply_gap_score'] >= 60:
        reasons.append("Room for significant expansion")
    
    if row['interest_score'] >= 80:
        reasons.append("Very high search interest")
    elif row['interest_score'] >= 60:
        reasons.append("Growing awareness and interest")
    
    if row['disposable_income'] >= 30000:
        reasons.append("High income market")
    
    for i, reason in enumerate(reasons, 1):
        print(f"     {i}. {reason}")

# ============================================================================
# STEP 6: Save results
# ============================================================================

# Save full results to CSV
#df_final.to_csv('opportunity_index_results.csv', index=False)

# Create a summary table for presentation
summary = df_final[[
    'rank', 'city', 'opportunity_index', 
    'potential_padel_players', 'padel_court_count', 'courts_per_100k'
]].sort_values('rank')

#summary.to_csv('opportunity_index_summary.csv', index=False)

print("\n" + "="*90)
print("FILES SAVED:")
print("="*90)
print("  ✓ opportunity_index_results.csv      (Full dataset with all scores)")
print("  ✓ opportunity_index_summary.csv      (Summary for presentation)")
print("="*90)

# ============================================================================
# STEP 7: Quick statistics
# ============================================================================

print("\n" + "="*90)
print("QUICK STATISTICS")
print("="*90)
print(f"  Average Opportunity Index:     {df_final['opportunity_index'].mean():.1f}")
print(f"  Highest Opportunity Index:     {df_final['opportunity_index'].max():.1f} ({df_final.loc[df_final['opportunity_index'].idxmax(), 'city']})")
print(f"  Total Potential Players:       {df_final['potential_padel_players'].sum():,}")
print(f"  Total Existing Courts:         {df_final['padel_court_count'].sum()}")
print(f"  Average Courts per 100k:       {df_final['courts_per_100k'].mean():.2f}")
print("="*90)
```

```python
padel_tennis_counts["city"] = padel_tennis_counts["city"].replace({
    "Munich": "München",
    "Frankfurt": "Frankfurt am Main"
})

# Merge all data together
merged = (
    pop_income_state
    .merge(germany_gtrend, on="state", how="left")
    .merge(padel_tennis_counts, on="city", how="left")
    .merge(gym_members, on="city", how="left")
    .merge(city_football_tennis, on="city", how="left")
    .merge(percentage_sports_population[['state', 'membership_rate_pct']], on="state", how="left")
)

print("Data loaded successfully!")
print(f"Analyzing {len(merged)} German cities\n")

# ============================================================================
# STEP 2: Calculate Opportunity Index
# ============================================================================

def calculate_opportunity_index(df):
    """
    Calculate opportunity index based on:
    - Potential padel players (from tennis, gym, football, newcomers)
    - Supply gap (courts per capita)
    - Search interest (Google Trends)
    
    CORRECTED Conversion rates (% who will TRY padel):
    - 15% of tennis players will try padel
    - 12% of gym members will try padel
    - 7% of football players will try padel
    - 2% of general population (newcomers)
    
    Note: These rates reflect realistic adoption, not the composition
    of the padel player base.
    """
    
    # Make a copy to avoid modifying original
    df = df.copy()
    
    # Fill missing values with 0 to avoid errors
    df['tennis_players'] = df['tennis_players'].fillna(0)
    df['gym_members'] = df['gym_members'].fillna(0)
    df['football_players'] = df['football_players'].fillna(0)
    df['padel_court_count'] = df['padel_court_count'].fillna(0)
    df['trend_score_five_years'] = df['trend_score_five_years'].fillna(0)
    
    # -------------------------------------------------------------------------
    # A) DEMAND SCORE: Calculate potential padel players
    # -------------------------------------------------------------------------
    
    # More realistic conversion rates:
    # - 15% of tennis players likely to try padel (mid-range estimate)
    # - 12% of gym members likely to try padel (mid-range estimate)
    # - 7% of football players likely to try padel (mid-range estimate)
    # - 2% of general population as newcomers
    
    df['potential_padel_players'] = (
        df['tennis_players'] * 0.15 +      # 15% of tennis players
        df['gym_members'] * 0.12 +          # 12% of gym members
        df['football_players'] * 0.07 +     # 7% of football players
        df['population'] * 0.02              # 2% newcomers from general pop
    )
    
    # Round and convert to int (now safe because we filled NaN)
    df['potential_padel_players'] = df['potential_padel_players'].round(0).astype(int)
    
    # Normalize to 0-100 scale (handle case where all values are 0)
    max_potential = df['potential_padel_players'].max()
    if max_potential > 0:
        df['demand_score'] = (
            (df['potential_padel_players'] / max_potential) * 100
        ).round(1)
    else:
        df['demand_score'] = 0
    
    # -------------------------------------------------------------------------
    # B) SUPPLY GAP SCORE: Fewer courts per capita = higher opportunity
    # -------------------------------------------------------------------------
    
    # Calculate courts per 100k population (avoid division by zero)
    df['courts_per_100k'] = (
        (df['padel_court_count'] / df['population']) * 100000
    ).replace([float('inf'), float('-inf')], 0).round(2)
    
    # Invert the score: fewer courts = higher opportunity
    max_density = df['courts_per_100k'].max()
    
    if max_density > 0:
        df['supply_gap_score'] = (
            100 - ((df['courts_per_100k'] / max_density) * 100)
        ).round(1)
    else:
        df['supply_gap_score'] = 100  # If no courts exist, max opportunity
    
    # -------------------------------------------------------------------------
    # C) INTEREST SCORE: Google Trends (already 0-100)
    # -------------------------------------------------------------------------
    
    df['interest_score'] = df['trend_score_five_years']
    
    # -------------------------------------------------------------------------
    # D) SPORTS CULTURE SCORE: Membership rate as indicator
    # -------------------------------------------------------------------------
    
    # Normalize membership_rate_pct to 0-100 scale
    # Higher membership rate = stronger sports culture = higher score
    df['membership_rate_pct'] = df['membership_rate_pct'].fillna(df['membership_rate_pct'].mean())
    
    max_membership = df['membership_rate_pct'].max()
    if max_membership > 0:
        df['sports_culture_score'] = (
            (df['membership_rate_pct'] / max_membership) * 100
        ).round(1)
    else:
        df['sports_culture_score'] = 0
    
    # -------------------------------------------------------------------------
    # E) FINAL OPPORTUNITY INDEX (weighted average)
    # -------------------------------------------------------------------------
    
    # Updated weights to include sports culture:
    # - 37.5% Demand (potential players)
    # - 37.5% Supply Gap (room for new courts)
    # - 15% Interest (search trends)
    # - 10% Sports Culture (organized sports participation)
    
    df['opportunity_index'] = (
        df['demand_score'] * 0.375 +
        df['supply_gap_score'] * 0.375 +
        df['interest_score'] * 0.15 +
        df['sports_culture_score'] * 0.10
    ).round(1)
    
    # Rank cities by opportunity
    df['rank'] = df['opportunity_index'].rank(ascending=False, method='min').astype(int)
    
    return df


# ===========================nb
=================================================
# STEP 3: Apply the calculation
# ============================================================================

df_final = calculate_opportunity_index(merged)

# ============================================================================
# STEP 4: Display results
# ============================================================================

# Select columns for display
result_columns = [
    'rank',
    'city',
    'state',
    'opportunity_index',
    'demand_score',
    'supply_gap_score',
    'interest_score',
    'sports_culture_score',
    'membership_rate_pct',
    'potential_padel_players',
    'padel_court_count',
    'courts_per_100k',
    'population',
    'disposable_income'
]

# Sort by opportunity index
results = df_final[result_columns].sort_values('opportunity_index', ascending=False)

print("="*90)
print("PADEL EXPANSION OPPORTUNITY INDEX - GERMAN CITIES")
print("="*90)
print(results.to_string(index=False))

# ============================================================================
# STEP 5: Show detailed insights for top 3 cities
# ============================================================================

print("\n" + "="*90)
print("TOP 3 EXPANSION OPPORTUNITIES - DETAILED ANALYSIS")
print("="*90)

top_3 = df_final.nlargest(3, 'opportunity_index')

for idx, row in top_3.iterrows():
    print(f"\n{'='*90}")
    print(f"RANK #{int(row['rank'])} - {row['city'].upper()}")
    print(f"{'='*90}")
    print(f"  Opportunity Index:        {row['opportunity_index']}/100")
    print(f"")
    print(f"  📊 SCORES:")
    print(f"     Demand Score:          {row['demand_score']}/100")
    print(f"     Supply Gap Score:      {row['supply_gap_score']}/100")
    print(f"     Interest Score:        {row['interest_score']}/100")
    print(f"     Sports Culture Score:  {row['sports_culture_score']}/100 ({row['membership_rate_pct']:.1f}% club membership)")
    print(f"")
    print(f"  👥 MARKET SIZE:")
    print(f"     Population:            {row['population']:,}")
    print(f"     Potential Players:     {row['potential_padel_players']:,}")
    print(f"     Disposable Income:     €{row['disposable_income']:,}")
    print(f"")
    print(f"  🏓 CURRENT SUPPLY:")
    print(f"     Existing Courts:       {row['padel_court_count']}")
    print(f"     Courts per 100k:       {row['courts_per_100k']}")
    print(f"")
    print(f"  💡 WHY THIS CITY:")
    
    # Generate insights
    reasons = []
    if row['demand_score'] >= 80:
        reasons.append(f"Very high demand ({int(row['potential_padel_players']):,} potential players)")
    elif row['demand_score'] >= 60:
        reasons.append(f"Strong demand ({int(row['potential_padel_players']):,} potential players)")
    
    if row['supply_gap_score'] >= 80:
        reasons.append("Severely undersupplied with courts")
    elif row['supply_gap_score'] >= 60:
        reasons.append("Room for significant expansion")
    
    if row['interest_score'] >= 80:
        reasons.append("Very high search interest")
    elif row['interest_score'] >= 60:
        reasons.append("Growing awareness and interest")
    
    if row['disposable_income'] >= 30000:
        reasons.append("High income market")
    
    if row['sports_culture_score'] >= 80:
        reasons.append(f"Very strong sports culture ({row['membership_rate_pct']:.1f}% club membership)")
    elif row['sports_culture_score'] >= 60:
        reasons.append(f"Active sports culture ({row['membership_rate_pct']:.1f}% club membership)")
    
    for i, reason in enumerate(reasons, 1):
        print(f"     {i}. {reason}")

# ============================================================================
# STEP 6: Save results
# ============================================================================

# Save full results to CSV
df_final.to_csv('opportunity_index_results.csv', index=False)

# Create a summary table for presentation
summary = df_final[[
    'rank', 'city', 'opportunity_index', 
    'potential_padel_players', 'padel_court_count', 'courts_per_100k'
]].sort_values('rank')

summary.to_csv('opportunity_index_summary.csv', index=False)

print("\n" + "="*90)
print("FILES SAVED:")
print("="*90)
print("  ✓ opportunity_index_results.csv      (Full dataset with all scores)")
print("  ✓ opportunity_index_summary.csv      (Summary for presentation)")
print("="*90)

# ============================================================================
# STEP 7: Quick statistics
# ============================================================================

print("\n" + "="*90)
print("QUICK STATISTICS")
print("="*90)
print(f"  Average Opportunity Index:     {df_final['opportunity_index'].mean():.1f}")
print(f"  Highest Opportunity Index:     {df_final['opportunity_index'].max():.1f} ({df_final.loc[df_final['opportunity_index'].idxmax(), 'city']})")
print(f"  Total Potential Players:       {df_final['potential_padel_players'].sum():,}")
print(f"  Total Existing Courts:         {df_final['padel_court_count'].sum()}")
print(f"  Average Courts per 100k:       {df_final['courts_per_100k'].mean():.2f}")
print("="*90)
```
